# Summer Olympics Data Analysis Assignment

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('summer.csv')
df
```

## 1. In how many cities Summer Olympics has been held so far?

```python
df.City.nunique()
```

## 2. Which sports are having the most number of Gold Medals won so far? (Top 5)

```python
gold = df[df['Medal'] == 'Gold']

gold_medals = gold.groupby(by='Sport')['Medal'].count().sort_values(ascending=False).reset_index()

plt.figure(figsize = (8,5))
sns.barplot(x='Sport', y='Medal', data=gold_medals.head(5))
plt.xticks(size=15)
plt.yticks(size=15)
plt.xlabel('Sport', size=20)
plt.ylabel('Gold Medals', size=20)
plt.show()
```

## 3. Which sports are having the most number of Medals won so far? (Top 5)

```python
medals = df.groupby(by='Sport')['Medal'].count().sort_values(ascending=False).reset_index()

plt.figure(figsize = (8,5))
sns.barplot(x='Sport', y='Medal', data=medals.head(5))
plt.xticks(size=15)
plt.yticks(size=15)
plt.xlabel('Sport', size=20)
plt.ylabel('Medals', size=20)
plt.show()
```

## 4. Which players have won the most number of Medals? (Top 5)

```python
players = df.groupby(by='Athlete')['Medal'].count().sort_values(ascending=False).reset_index()

plt.figure(figsize = (10,5))
sns.barplot(x='Athlete', y='Medal', data=players.head(5))
plt.xticks(rotation=45, size=15)
plt.yticks(size=15)
plt.xlabel('Athlete', size=20)
plt.ylabel('Medals', size=20)
plt.show()
```

## 5. Which players have won the most number of Gold Medals? (Top 5)

```python
players_gold = gold.groupby(by='Athlete')['Medal'].count().sort_values(ascending=False).reset_index()

plt.figure(figsize = (10,5))
sns.barplot(x='Athlete', y='Medal', data=players_gold.head(5))
plt.xticks(rotation=45, size=15)
plt.yticks(size=15)
plt.xlabel('Athlete', size=20)
plt.ylabel('Gold Medals', size=20)
plt.show()
```

## 6. In which year did India win the first Gold Medal in Summer Olympics?

```python
india = df[df['Country'] == 'IND']
gold_india = india[india['Medal'] == 'Gold']
gold_india.Year.unique()[0]
```

## 7. Which events are the most popular in terms of the number of players? (Top 5)

```python
events = df.groupby(by='Event')['Athlete'].count().sort_values(ascending=False).reset_index()

plt.figure(figsize = (8,5))
sns.barplot(x='Event', y='Athlete', data=events.head(5))
plt.xticks(rotation=45, size=15)
plt.yticks(size=15)
plt.xlabel('Event', size=20)
plt.ylabel('Players', size=20)
plt.show()
```

## 8. Which sports are having the most female Gold Medalists? (Top 5)

```python
female = gold[gold['Gender'] == 'Women']

gold_women = female.groupby(by='Sport')['Medal'].count().sort_values(ascending=False).reset_index()

plt.figure(figsize = (8,5))
sns.barplot(x='Sport', y='Medal', data=gold_women.head(5))
plt.xticks(size=15)
plt.yticks(size=15)
plt.xlabel('Sport', size=20)
plt.ylabel('Medals', size=20)
plt.show()
```
