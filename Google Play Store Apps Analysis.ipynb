```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
apps = pd.read_csv('googleplaystore.csv')
apps
```

```python
apps.info()
```

```python
apps.drop(['Last Updated', 'Current Ver', 'Android Ver'], axis=1, inplace=True)
```

```python
apps.shape
```

```python
apps['Category'].unique()
```

```python
apps[apps['Category'] == '1.9']
```

```python
apps.loc[10472, 'Category'] = 'LIFESTYLE'
apps.loc[10472, 'Rating'] = 1.9
apps.loc[10472, 'Reviews'] = '19.0'
apps.loc[10472, 'Size'] = '3.0M'
apps.loc[10472, 'Installs'] = '1,000+'
apps.loc[10472, 'Type'] = 'Free'
apps.loc[10472, 'Price'] = '0'
apps.loc[10472, 'Content Rating'] = 'Everyone'
apps.loc[10472, 'Genres'] = 'Lifestyle'
apps.loc[10472]
```

```python
apps.duplicated().sum()
```

```python
apps.drop_duplicates(inplace=True)
```

```python
apps.duplicated(subset='App').sum()
```

```python
apps.drop_duplicates(subset='App', keep='first', inplace = True)
```

```python
apps.dropna(inplace=True)
```

```python
apps
```

```python
apps['Price'].unique()
```

```python
apps['Price'] = apps.Price.str.replace("$","").astype(float)
apps['Price'].dtype
```

```python
apps['Reviews'].unique()
```

```python
apps['Reviews'] = apps['Reviews'].replace("0.0", 0)
apps['Reviews'] = apps['Reviews'].astype(float)
apps['Reviews'].dtype
```

```python
apps['Size'].unique()
```

```python
# convert all size values to the KiloByte type.
apps['Size'] = apps['Size'].replace("Varies with device", np.nan)
apps['Size'] = apps['Size'].str.replace("M", "000") 
apps['Size'] = apps['Size'].str.replace("k", "")
apps['Size'] = apps['Size'].astype(float)
apps['Size'].dtype
```

```python
apps['Installs'].unique()
```

```python
apps['Installs'] = apps['Installs'].str.replace(",", "")
apps['Installs'] = apps['Installs'].str.replace("+", "")
apps['Installs'] = apps['Installs'].astype(float)
apps['Installs'].dtype
```

```python
apps['Installs'] = apps['Installs'].apply(lambda x: int(x))
apps[['App','Installs','Category','Content Rating','Price']].head()
new = apps[['App','Installs','Category','Content Rating','Price','Reviews','Genres']].groupby(['Installs'], sort=True)
new = new.filter(lambda x: x['Installs'].mean() >= 1000000000)
new = new.sort_values(by=['Installs'])
print("Most Installed Free Apps:")
new
```

```python
newdf = apps[['App','Category','Content Rating','Price','Reviews']].groupby(['Price'], sort=True)
newdf = newdf.filter(lambda x: x['Price'].mean() != 0)
newdf = newdf.sort_values(by=['Price'])

newdf2 = apps[['App','Installs','Genres']].groupby(['Installs'], sort = True)
newdf2 = newdf2.filter(lambda x: x['Installs'].mean() >= 1000000)
newdf2 = newdf2.sort_values(by=['Installs'])

s1 = pd.merge(newdf, newdf2, how='inner', on=['App'])
print("Most Installed Paid Apps:")
s1
```

```python
apps['Rating'] = apps['Rating'].apply(lambda x: float(x))
apps['Reviews'] = apps['Reviews'].apply(lambda x: int(x))

df_rate = apps[['App','Rating','Category','Content Rating']].groupby(['Rating'], sort=True)
df_rate = df_rate.filter(lambda x: x['Rating'].mean() >= 4.5)
df_rate = df_rate.sort_values(by=['Rating'])

df_reviews = apps[['App','Reviews']].groupby(['Reviews'], sort=True)
df_reviews = df_reviews.filter(lambda x: x['Reviews'].mean() >= 255435)
df_reviews = df_reviews.sort_values(by=['Reviews'])

df_installs = apps[['App','Installs']].groupby(['Installs'], sort=True)
df_installs = df_installs.filter(lambda x: x['Installs'].mean() >= 10000000)
df_installs = df_installs.sort_values(by=['Installs'])

s1 = pd.merge(df_reviews, df_rate, how='inner', on=['App'])
s2 = pd.merge(s1, df_installs, how='inner', on=['App'])
print("Best Apps:")
s2
```

```python
plt.figure(figsize = (10,10))
likeable = sns.countplot(y="Category", data=s2, palette = "Set3")
likeable
plt.title('Most Likeable App Category', size = 20)
plt.show()
```
